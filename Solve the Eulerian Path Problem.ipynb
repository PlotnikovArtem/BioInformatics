CODE CHALLENGE: Solve the Eulerian Path Problem.
     
     Input: The adjacency list of a directed graph that has an Eulerian path.
     
     Output: An Eulerian path in this graph.

Extra Dataset

Sample Input:

0 -> 2

1 -> 3,4

10 -> 12

11 -> 2

12 -> 11

13 -> 15

14 -> 13

15 -> 2

2 -> 1,10,14

3 -> 0

4 -> 5

5 -> 6

6 -> 1

7 -> 2

9 -> 7

```python
def explore(elem, edges, nodes_w_edges, path):
    path.append(elem)
    while cur_pos[elem] < out_deg[elem]:
        if cur_pos[elem] + 1 >= out_deg[elem]:
            if elem in nodes_w_edges:
                del nodes_w_edges[elem]
        else:
            nodes_w_edges[elem] = len(path) - 1
        cur_pos[elem] += 1
        path.append(conf[elem][cur_pos[elem] - 1])
        elem = conf[elem][cur_pos[elem] - 1] 
        edges -= 1
    return edges, nodes_w_edges, path

def EulerianCycle(nodes_w_edges, edges, path):
    for var in range(n):
        if in_deg[var] != out_deg[var]:
            if in_deg[var] >= out_deg[var]:
                unbalanced.insert(0, var)
            else:
                unbalanced.append(var)
    if len(unbalanced) > 0:
        conf[unbalanced[0]].append(unbalanced[1])
        out_deg[unbalanced[0]] += 1
        in_deg[unbalanced[1]] += 1
    edges, nodes_w_edges, path = explore(0, edges, nodes_w_edges, path)
    while edges > 0:
        node, startPos = nodes_w_edges.popitem()
        l = len(path) - 1
        path = path[startPos:l] + path[:startPos]
        for knot, pos in nodes_w_edges.items():
            if pos < startPos:
                nodes_w_edges[knot] = pos + l - startPos
            else:
                nodes_w_edges[knot] = pos - startPos
        edges, nodes_w_edges, path = explore(node, edges, nodes_w_edges, path)
    if len(unbalanced) > 0:
        for i in range(len(path) - 1):
            if path[i + 1] == unbalanced[1]:
                if path[i] == unbalanced[0]:
                    position = i + 1
                    l = len(path) - 1
                    path = path[position:l] + path[:position]
                    for knot, pos in nodes_w_edges.items():
                        if pos < position:
                            nodes_w_edges[knot] = pos + l - position
                        else:
                            nodes_w_edges[knot] = pos - position
                    break
    return path          

edges = 0
nodes_w_edges = dict()
path = []
unbalanced = []

text = input()
data = text.split(' ')
var_max = 0
for i in range(len(data) // 3):
    var_max = max(int(data[i*3]), var_max, max(list(map(int, data[i*3+2].split(',')))))
n = var_max + 1
conf = [[]] * n
in_deg = [0] * n
out_deg = [0] * n
cur_pos = [0] * n
for i in range(len(data) // 3):
    count = int(data[i*3])
    conf[count] = list(map(int, data[i*3+2].split(',')))
    for var in conf[count]:
        in_deg[var] += 1
    l = len(conf[count])
    out_deg[count] = l
    edges += l
            
path = EulerianCycle(nodes_w_edges, edges, path)

for i in range(len(path)):
    if i == len(path) - 1:
        print(path[i])
        continue
    print(path[i],'->', sep='',end= '')
```
